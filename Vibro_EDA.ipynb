```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Ensure your DataFrame 'df' is already set up and contains the correct data

# Список столбцов для корреляционной матрицы
columns_to_correlate = [
    "ORB_MAX_EJE", "AMP_EJE_X", "PHI_EJE_X", "AMP_EJE_Y", "PHI_EJE_Y",
    "RMS_AV_EJE_X", "RMS_MAX_EJE_X", "PP_AV_EJE_X", "PP_MAX_EJE_X",
    "RMS_AV_EJE_Y", "RMS_MAX_EJE_Y", "PP_AV_EJE_Y", "PP_MAX_EJE_Y"
]
columns_to_correlate = [
   "H","VEL_GEN_AV","PAV", "AMP_EJE_X", "PHI_EJE_X", "AMP_EJE_Y", "PHI_EJE_Y"
]

# Создаем корреляционную матрицу
correlation_matrix = df[columns_to_correlate].corr()

# Создаем маску для треугольника корреляционной матрицы, чтобы отобразить только нижнюю половину
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Создаем фигуру с заданными размерами
plt.figure(figsize=(14, 12))

# Назначаем заголовок графика
plt.title('Correlation Heatmap', fontsize=18)

# Создаем тепловую карту
sns.heatmap(correlation_matrix,
            mask=mask,
            cmap='coolwarm',  # Two-tone gradient for better visibility
            annot=True,       # Annotate with text
            fmt=".2f",        # Format the annotations to 2 decimal places
            linewidths=0.5,   # Width of the lines that divide each cell
            cbar_kws={"shrink": .8}, # Shrink the color bar to better fit the visualization
            square=True)      # Ensures that each cell is square-shaped

# Показываем график
plt.show()
```
