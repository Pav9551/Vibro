```python
import pandas as pd

# Укажите путь к файлу
file_path_times = "U3_times.csv"
# Загружаем файл в DataFrame
df_times = pd.read_csv(file_path_times, sep=";", header=None)
# Удаляем последнюю колонку
df_times = df_times.drop(df_times.columns[-1], axis=1)

# Выводим DataFrame, чтобы убедиться, что колонка удалена
times = df_times.iloc[0].values
times
```

```python
# Укажите путь к файлу
file_path_yValues = "U3_yValues.csv"
# Загружаем файл в DataFrame
df_yValues = pd.read_csv(file_path_yValues, sep=";", header=None)
# Удаляем последнюю колонку
df_yValues = df_yValues.drop(df_yValues.columns[-1], axis=1)
df_yValues
```

```python
# Получаем первую строку как массив
N = 187
array_x = df_xValues.iloc[N].values

# Выводим результат
df_x = pd.DataFrame({'times': times, 'array': array_x[3:]})
# Заменяем запятую на точку в строках и затем преобразуем в вещественные числа
df_x['times'] = df_x['times'].str.replace(',', '.').astype(float)
df_x['array'] = df_x['array'].str.replace(',', '.').astype(float)
df_x
```

```python
array_y = df_yValues.iloc[N].values

# Выводим результат
df_y = pd.DataFrame({'times': times, 'array': array_y[3:]})
# Заменяем запятую на точку в строках и затем преобразуем в вещественные числа
df_y['times'] = df_y['times'].str.replace(',', '.').astype(float)
df_y['array'] = df_y['array'].str.replace(',', '.').astype(float)
df_y
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq

def filter_signal(df):
    # Параметры
    fs = 1 / (df['times'][1] - df['times'][0])  # Частота дискретизации (Гц)
    t = df['times'].values
    signal = df['array'].values

    # Преобразование Фурье
    N = len(signal)
    fft_values = fft(signal)
    freqs = fftfreq(N, 1/fs)

    # Фильтрация: оставляем только первую гармонику
    fft_filtered = np.zeros_like(fft_values)
    dominant_freq_index = np.argmax(np.abs(fft_values[:N//2]))  # Индекс основной частоты
    fft_filtered[dominant_freq_index] = fft_values[dominant_freq_index]
    fft_filtered[-dominant_freq_index] = fft_values[-dominant_freq_index]  # Для симметричной фильтрации (отрицательная частота)

    # Обратное преобразование Фурье (получаем отфильтрованный сигнал)
    filtered_signal = np.real(ifft(fft_filtered))
    return t , signal, filtered_signal
```

```python
# Применение фильтра для df_x и df_y
t, signal_x, filtered_signal_x = filter_signal(df_x)
t, signal_y, filtered_signal_y = filter_signal(df_y)

# Построение графиков
plt.figure(figsize=(10, 12))

# График для df_x: исходный и отфильтрованный сигнал
plt.subplot(3, 1, 1)
plt.plot(t, signal_x, label='Исходный сигнал (df_x)', color='blue')
plt.plot(t, filtered_signal_x, label='Отфильтрованный сигнал (df_x)', color='red')
plt.title('Сигнал для df_x: до и после фильтрации')
plt.xlabel('Время (сек)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid()

# График для df_y: исходный и отфильтрованный сигнал
plt.subplot(3, 1, 2)
plt.plot(t, signal_y, label='Исходный сигнал (df_y)', color='green')
plt.plot(t, filtered_signal_y, label='Отфильтрованный сигнал (df_y)', color='orange')
plt.title('Сигнал для df_y: до и после фильтрации')
plt.xlabel('Время (сек)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid()

# Объединенный график после фильтрации: оба отфильтрованных сигнала
plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal_x, label='Отфильтрованный сигнал (df_x)', color='red')
plt.plot(t, filtered_signal_y, label='Отфильтрованный сигнал (df_y)', color='orange')
plt.title('Объединенный сигнал после фильтрации')
plt.xlabel('Время (сек)')
plt.ylabel('Амплитуда')
plt.legend()
plt.grid()

# Показать графики
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import correlate

# Пример данных (замени на свои точки)
x_data = np.linspace(0, 10, 100)
y_data = 3 * np.sin(2 * x_data + 1) + 2 + np.random.normal(0, 0.2, size=len(x_data))

# Определение целевой функции
def sinusoid(x, A, B, C, D):
    return A * np.sin(B * x + C) + D

# Автокорреляция для нахождения периода
corr = correlate(y_data - np.mean(y_data), y_data - np.mean(y_data), mode='full')
corr = corr[len(corr)//2:]
peaks = np.where((corr[1:-1] > corr[:-2]) & (corr[1:-1] > corr[2:]))[0]
if len(peaks) > 1:
    T_guess = x_data[peaks[1]] - x_data[peaks[0]]
else:
    T_guess = x_data[-1] - x_data[0]
B_guess = 2 * np.pi / T_guess

# Определение фазового сдвига
C_guess = x_data[np.argmax(y_data)] if len(peaks) > 0 else 0

# Начальные приближения
A_guess = (np.max(y_data) - np.min(y_data)) / 2
D_guess = np.mean(y_data)

initial_guess = [A_guess, B_guess, C_guess, D_guess]

# Подгонка
params, covariance = curve_fit(sinusoid, x_data, y_data, p0=initial_guess)
A_fit, B_fit, C_fit, D_fit = params

# Генерация аппроксимирующей кривой
y_fit = sinusoid(x_data, *params)

# Визуализация
plt.scatter(x_data, y_data, label="Исходные данные", color='blue', alpha=0.5)
plt.plot(x_data, y_fit, label=f"Подгонка: {A_fit:.2f}sin({B_fit:.2f}x + {C_fit:.2f}) + {D_fit:.2f}", color='red')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Аппроксимация синусоидальной функции")
plt.show()

# Вывод найденных параметров
print(f"Аппроксимирующая функция: y = {A_fit:.2f} * sin({B_fit:.2f} * x + {C_fit:.2f}) + {D_fit:.2f}")
```
